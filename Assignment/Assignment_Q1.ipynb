```python
import pandas as pd
import numpy as np
```

```python
fk = pd.read_csv('Flipkart-Laptops.csv')
fk['Company'] = fk['Product Name'].str.split().str[0]
fk['Actual price'] = fk['Actual price'].str.replace('?', '', regex=False).str.replace(',', '').str.strip()
fk['Discount price'] = fk['Discount price'].str.replace('?', '', regex=False).str.replace(',', '').str.strip()
fk['Actual price'] = pd.to_numeric(fk['Actual price'], errors='coerce')
fk['Discount price'] = pd.to_numeric(fk['Discount price'], errors='coerce')

fk
```

```python
fk['Rating'] = fk['Rating'].replace('NIL', np.nan)
fk['Rating'] = fk['Rating'].str.replace(' Ratings', '', regex=False).str.replace(',', '', regex=False)
fk['Rating'] = pd.to_numeric(fk['Rating'], errors='coerce')
```

```python
count_over_3000_ratings = fk[fk['Rating'] > 3000].shape[0]
count_over_3000_ratings
```

```python
fk['Reviews'] = fk['Reviews'].replace('NIL', np.nan)
fk['Reviews'] = fk['Reviews'].str.replace(' Reviews', '', regex=False)
fk['Reviews'] = pd.to_numeric(fk['Reviews'], errors='coerce')

fk['Stars'] = fk['Stars'].replace('NIL', np.nan)
fk['Stars'] = pd.to_numeric(fk['Stars'], errors='coerce')
```

```python
rating_cats = {
    'Poor (1-2)': fk[(fk['Stars'] >= 1) & (fk['Stars'] < 2)].shape[0],
    'Average (2-3)': fk[(fk['Stars'] >= 2) & (fk['Stars'] < 3)].shape[0],
    'Good (3-4)': fk[(fk['Stars'] >= 3) & (fk['Stars'] < 4)].shape[0],
    'Excellent (4-5)': fk[(fk['Stars'] >= 4) & (fk['Stars'] <= 5)].shape[0]
}
rating_cats
```
